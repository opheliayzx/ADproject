```python
import numpy as np
import skimage.io as io
from skimage.transform import rotate, AffineTransform, warp
from skimage.util import random_noise
from skimage.filters import gaussian 
from cv2 import LUT
import matplotlib.pyplot as plt
import data_augmentation 
import PIL.Image as Image
import torch
```

```python
img = np.load("/Users/ophelia/Desktop/BrainSamples/Brain2/Block1/Tau/L5/1636.npy")

plt.imshow(img[:,:,3])

img = torch.from_numpy(img)

sample = {"image": img, "label": img}
```

```python
# Rotations
rotated_sample = data_augmentation.augment_sample(sample, random_orientation=True)
plt.imshow(rotated_sample['image'][:,:,0:3]) 
```

```python
# # Gamma transform

gamma_sample = data_augmentation.augment_sample(sample, gamma=True)
plt.imshow(gamma_sample['image'][:,:,0:3]) 
```

```python
# avg_noise

avg_noise_sample = data_augmentation.augment_sample(sample, avg_noise=True)
plt.imshow(avg_noise_sample["image"][:,:,0:3])
```

```python
# total noise
total_noise_sample = data_augmentation.augment_sample(sample, total_noise=True)
plt.imshow(total_noise_sample["image"][:,:,0:3])
```

```python
# blur
blur_sample = data_augmentation.augment_sample(sample, blur=True)
plt.imshow(blur_sample["image"][:,:,0:3]/256)
```
